<p><b>–ú–µ—Ç–∞:</b> –∑–∞—Å–≤–æ—ó—Ç–∏ —Ç–µ—Ö–Ω–æ–ª–æ–≥—ñ—é —Ä–µ–∞–ª—ñ–∑–∞—Ü—ñ—ó —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–∏—Ö –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤ –∑–∞—Å–æ–±–∞–º–∏ Python —ñ –æ—Ü—ñ–Ω—é–≤–∞–Ω–Ω—è —ó—Ö —Å–∫–ª–∞–¥–Ω–æ—Å—Ç—ñ –∑ –≤–∏–∫–æ—Ä–∏—Å—Ç–∞–Ω–Ω—è–º –≥–æ–ª–æ–≤–Ω–æ—ó —Ç–µ–æ—Ä–µ–º–∏ —Ä–µ–∫—É—Ä—Å—ñ—ó.</p>

<h5 style="text-align:center;">–ü–æ—Ä—è–¥–æ–∫ –≤–∏–∫–æ–Ω–∞–Ω–Ω—è —Ä–æ–±–æ—Ç–∏</h5>

<p>1. –†–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ –ø—Ä–æ—Ü–µ–¥—É—Ä—É –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —Ñ–∞–∫—Ç–æ—Ä—ñ–∞–ª—É –∑–∞ –¥–æ–ø–æ–º–æ–≥–æ—é —Ü–∏–∫–ª—É for. –û—Ü—ñ–Ω–∏—Ç–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É.</p>

```python
def factorial(n):
    result = 1
    for i in range(1, n + 1):
        result *= i
    return result

print(factorial(5))
```

```python
def fac(n):
    if n == 0:
        return 1
    return fac(n - 1) * n

n = int(input("–í–≤–µ–¥—ñ—Ç—å —á–∏—Å–ª–æ: "))
print(fac(n))
```

–û—Ç–∂–µ, –∑–∞–≥–∞–ª—å–Ω–∞ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –∞–ª–≥–æ—Ä–∏—Ç–º—É ‚Äî O(n).

<p>2. –û—Ü—ñ–Ω–∏—Ç–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –æ–±—á–∏—Å–ª–µ–Ω–Ω—è —Ñ–∞–∫—Ç–æ—Ä—ñ–∞–ª—É.</p>
<ul>
    <li>–ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏–∫–ª–∏–∫—ñ–≤ —Ñ—É–Ω–∫—Ü—ñ—ó: –ö—ñ–ª—å–∫—ñ—Å—Ç—å —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–∏—Ö –≤–∏–∫–ª–∏–∫—ñ–≤ –ø—Ä—è–º–æ –ø—Ä–æ–ø–æ—Ä—Ü—ñ–π–Ω–∞ –∑–Ω–∞—á–µ–Ω–Ω—é n, –æ—Å–∫—ñ–ª—å–∫–∏ –º–∏ –∑–º–µ–Ω—à—É—î–º–æ n –Ω–∞ 1 –Ω–∞ –∫–æ–∂–Ω–æ–º—É –∫—Ä–æ—Ü—ñ, –∞–∂ –¥–æ –¥–æ—Å—è–≥–Ω–µ–Ω–Ω—è n=0.</li>
    <li>–ß–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∫–æ–∂–Ω–æ–≥–æ –≤–∏–∫–ª–∏–∫—É: –ö–æ–∂–µ–Ω –≤–∏–∫–ª–∏–∫ –≤–∏–∫–æ–Ω—É—î –æ–¥–Ω—É –æ–ø–µ—Ä–∞—Ü—ñ—é (–º–Ω–æ–∂–µ–Ω–Ω—è) —ñ –∑–∞–π–º–∞—î –ø–æ—Å—Ç—ñ–π–Ω–∏–π —á–∞—Å O(1).</li>
</ul>
–û—Ç–∂–µ, –∑–∞–≥–∞–ª—å–Ω–∞ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏–∫–ª–∏–∫—ñ–≤ ‚Äî n, –∞ –∫–æ–∂–µ–Ω –≤–∏–∫–ª–∏–∫ –∑–∞–π–º–∞—î —á–∞—Å O(1).

```python
def fibonacci(n):
    if n == 0:
        return 0
    if n in (1, 2):
        return 1
    return fibonacci(n - 1) + fibonacci(n - 2)

print(fibonacci(30))
```

<p>3. –û—Ü—ñ–Ω–∏—Ç–∏ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –æ–±—á–∏—Å–ª–µ–Ω–Ω—è ùëõ-–≥–æ —á–∏—Å–ª–∞ –§—ñ–±–æ–Ω–∞—á—á—ñ.</p>
<ul>
    <li>–ö—ñ–ª—å–∫—ñ—Å—Ç—å —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–∏—Ö –≤–∏–∫–ª–∏–∫—ñ–≤: –ö—ñ–ª—å–∫—ñ—Å—Ç—å –≤–∏–∫–ª–∏–∫—ñ–≤ –∑—Ä–æ—Å—Ç–∞—î –µ–∫—Å–ø–æ–Ω–µ–Ω—Ü—ñ–∞–ª—å–Ω–æ –∑ –∫–æ–∂–Ω–∏–º –∑–º–µ–Ω—à–µ–Ω–Ω—è–º n. –ö–æ–∂–µ–Ω —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–∏–π –≤–∏–∫–ª–∏–∫ –¥–ª—è n –ø—Ä–∏–∑–≤–æ–¥–∏—Ç—å –¥–æ –¥–≤–æ—Ö –¥–æ–¥–∞—Ç–∫–æ–≤–∏—Ö –≤–∏–∫–ª–∏–∫—ñ–≤: –æ–¥–Ω–æ–≥–æ –¥–ª—è n‚àí1 —ñ —ñ–Ω—à–æ–≥–æ –¥–ª—è n‚àí2.</li>
    <li>–ß–∞—Å–æ–≤–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å: –ö—ñ–ª—å–∫—ñ—Å—Ç—å —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–∏—Ö –≤–∏–∫–ª–∏–∫—ñ–≤ –º–æ–∂–Ω–∞ –æ—Ü—ñ–Ω–∏—Ç–∏ —è–∫ O(2n), –æ—Å–∫—ñ–ª—å–∫–∏ –¥–ª—è –∫–æ–∂–Ω–æ–≥–æ —Ä—ñ–≤–Ω—è –º–∏ –º–∞—î–º–æ –ø—Ä–∏–±–ª–∏–∑–Ω–æ –ø–æ–¥–≤–æ—î–Ω–Ω—è –∫—ñ–ª—å–∫–æ—Å—Ç—ñ –≤–∏–∫–ª–∏–∫—ñ–≤. –¶–µ –≤—ñ–¥–±—É–≤–∞—î—Ç—å—Å—è —á–µ—Ä–µ–∑ –¥—É–±–ª—é–≤–∞–Ω–Ω—è –æ–±—á–∏—Å–ª–µ–Ω—å, –Ω–∞–ø—Ä–∏–∫–ª–∞–¥, –ø—Ä–∏ –æ–±—á–∏—Å–ª–µ–Ω–Ω—ñ F(n‚àí1) —ñ F(n‚àí2), –æ–±—á–∏—Å–ª–µ–Ω–Ω—è –¥–ª—è –º–µ–Ω—à–∏—Ö –∑–Ω–∞—á–µ–Ω—å —á–∞—Å—Ç–æ –ø–æ–≤—Ç–æ—Ä—é—é—Ç—å—Å—è.</li>
</ul>
–ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å
–ó–∞–≥–∞–ª—å–Ω–∞ –∞—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å —Ä–µ–∫—É—Ä—Å–∏–≤–Ω–æ–≥–æ –∞–ª–≥–æ—Ä–∏—Ç–º—É –æ–±—á–∏—Å–ª–µ–Ω–Ω—è n-–≥–æ —á–∏—Å–ª–∞ –§—ñ–±–æ–Ω–∞—á—á—ñ ‚Äî O(2 n). –¶–µ –µ–∫—Å–ø–æ–Ω–µ–Ω—Ü—ñ–∞–ª—å–Ω–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å, –æ—Å–∫—ñ–ª—å–∫–∏ –∫—ñ–ª—å–∫—ñ—Å—Ç—å –æ–ø–µ—Ä–∞—Ü—ñ–π –ø–æ–¥–≤–æ—é—î—Ç—å—Å—è –∑ –∫–æ–∂–Ω–∏–º –∑–±—ñ–ª—å—à–µ–Ω–Ω—è–º n.

```python
def merge(left, right):
    result = []
    i, j = 0, 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
    result += left[i:]
    result += right[j:]
    return result

def mergesort(lst):
    if len(lst) < 2:
        return lst
    middle = len(lst) // 2
    left = mergesort(lst[:middle])
    right = mergesort(lst[middle:])
    return merge(left, right)


a = [6, 5, 3, 1, 8, 7, 2, 4]
print(mergesort(a))
```
