<b>1. Що таке pip і conda та чим вони відрізняються між собою?</b>
<li><b>pip</b> — це стандартний пакетний менеджер для Python, який використовується для встановлення та управління Python-бібліотеками, що знаходяться в репозиторії PyPI.</li>
<li><b>conda</b> — це пакетний менеджер, який працює з будь-яким програмним забезпеченням, не тільки з Python. Він може встановлювати пакети і управляти середовищами як для Python, так і для інших мов програмування, і використовується в основному в середовищах Anaconda та Miniconda.</li>
<li>Основна відмінність: pip працює тільки з Python-пакетами, а conda підтримує пакети для різних мов і забезпечує повну ізоляцію середовища.</li>

<b>4. Написати процедуру на Python сортування масиву методом бульбашки.</b>

```python
import time

def bubble_sort(arr):
    def swap(i, j):
        arr[i], arr[j] = arr[j], arr[i]

    n= len(arr)
    swapped = True

    x = -1 
    while swapped:
        swapped = False
        x = x + 1
        for i in range(1, n-x):
            if arr[i - 1] > arr[i]:
                swap(i - 1, i)
                swapped = True
                
    return arr

my_array = [64, 34, 25, 12, 22, 11, 90]
start_time = time.time()
sorted_array = bubble_sort(my_array)
end_time = time.time() 

print("Відсортований масив:")
print("--- %s seconds ---" % (end_time - start_time))
```
