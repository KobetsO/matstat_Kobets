        <li>–ù–∞–π–∫—Ä–∞—â–∏–π –≤–∏–ø–∞–¥–æ–∫ (O(n)):<p>–£ –Ω–∞–π–∫—Ä–∞—â–æ–º—É –≤–∏–ø–∞–¥–∫—É (–∫–æ–ª–∏ –º–∞—Å–∏–≤ –≤–∂–µ –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π), –∞–ª–≥–æ—Ä–∏—Ç–º –ø—Ä–æ—Å—Ç–æ –ø—Ä–æ—Ö–æ–¥–∏—Ç—å –ø–æ –º–∞—Å–∏–≤—É —ñ –ø–æ—Ä—ñ–≤–Ω—é—î –∫–æ–∂–µ–Ω –µ–ª–µ–º–µ–Ω—Ç –∑ –ø–æ–ø–µ—Ä–µ–¥–Ω—ñ–º, –Ω–µ –∑–¥—ñ–π—Å–Ω—é—é—á–∏ –∑—Å—É–≤—ñ–≤. –£ —Ç–∞–∫–æ–º—É –≤–∏–ø–∞–¥–∫—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –±—É–¥–µ –ª—ñ–Ω—ñ–π–Ω–æ—é ‚Äî O(n).</p></li>

    </ul>
<p><strong>2. –°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –±—É–ª—å–±–∞—à–∫–æ—é (Bubble Sort)</strong></p>
<p><strong>–û–ø–∏—Å –∞–ª–≥–æ—Ä–∏—Ç–º—É:</strong> –°–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –±—É–ª—å–±–∞—à–∫–æ—é –ø—Ä–∞—Ü—é—î —à–ª—è—Ö–æ–º –±–∞–≥–∞—Ç–æ—Ä–∞–∑–æ–≤–æ–≥–æ –ø—Ä–æ—Ö–æ–¥–∂–µ–Ω–Ω—è –ø–æ –º–∞—Å–∏–≤—É, –ø–æ—Ä—ñ–≤–Ω—é—é—á–∏ —Å—É—Å—ñ–¥–Ω—ñ –µ–ª–µ–º–µ–Ω—Ç–∏ —ñ –º—ñ–Ω—è—é—á–∏ —ó—Ö –º—ñ—Å—Ü—è–º–∏, —è–∫—â–æ –≤–æ–Ω–∏ —Å—Ç–æ—è—Ç—å —É –Ω–µ–ø—Ä–∞–≤–∏–ª—å–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É. –ö–æ–∂–Ω–∞ —ñ—Ç–µ—Ä–∞—Ü—ñ—è "–≤—Å–ø–ª–∏–≤–∞—î" –Ω–∞–π–±—ñ–ª—å—à–∏–π –Ω–µ–≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π –µ–ª–µ–º–µ–Ω—Ç –¥–æ –∫—ñ–Ω—Ü—è –º–∞—Å–∏–≤—É, —Ç–æ–º—É –ø—ñ—Å–ª—è –∫–æ–∂–Ω–æ—ó —ñ—Ç–µ—Ä–∞—Ü—ñ—ó –∫—ñ–ª—å–∫—ñ—Å—Ç—å –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –¥–ª—è –ø–µ—Ä–µ–≤—ñ—Ä–∫–∏ –∑–º–µ–Ω—à—É—î—Ç—å—Å—è.</p>
<p>–ê—Å–∏–º–ø—Ç–æ—Ç–∏—á–Ω–∞ —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å:</p>
<ul>
    <li>–ù–∞–π–≥—ñ—Ä—à–∏–π –≤–∏–ø–∞–¥–æ–∫ (O(n¬≤)):<p>–£ –Ω–∞–π–≥—ñ—Ä—à–æ–º—É –≤–∏–ø–∞–¥–∫—É (–∫–æ–ª–∏ –º–∞—Å–∏–≤ –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π —É –∑–≤–æ—Ä–æ—Ç–Ω–æ–º—É –ø–æ—Ä—è–¥–∫—É) –∞–ª–≥–æ—Ä–∏—Ç–º –±—É–¥–µ –≤–∏–∫–æ–Ω—É–≤–∞—Ç–∏ n –ø–æ—Ä—ñ–≤–Ω—è–Ω—å –∑–∞ –∫–æ–∂–Ω—É —ñ—Ç–µ—Ä–∞—Ü—ñ—é —ñ –ø–µ—Ä–µ—Å—Ç–∞–≤–ª—è—Ç–∏ –µ–ª–µ–º–µ–Ω—Ç–∏ –Ω–∞ –∫–æ–∂–Ω–æ–º—É –∫—Ä–æ—Ü—ñ. –î–ª—è n –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ –ø–æ—Ç—Ä—ñ–±–Ω–æ –≤–∏–∫–æ–Ω–∞—Ç–∏ O(n2) –æ–ø–µ—Ä–∞—Ü—ñ–π.</p></li>
    <li>–ù–∞–π–∫—Ä–∞—â–∏–π –≤–∏–ø–∞–¥–æ–∫ (O(n)):<p>–£ –Ω–∞–π–∫—Ä–∞—â–æ–º—É –≤–∏–ø–∞–¥–∫—É (–∫–æ–ª–∏ –º–∞—Å–∏–≤ –≤–∂–µ –≤—ñ–¥—Å–æ—Ä—Ç–æ–≤–∞–Ω–∏–π), –∞–ª–≥–æ—Ä–∏—Ç–º –ø—Ä–æ—Ö–æ–¥–∏—Ç—å –º–∞—Å–∏–≤ –æ–¥–∏–Ω —Ä–∞–∑ —ñ –≤–∏—è–≤–ª—è—î, —â–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–∫–∏ –Ω–µ –ø–æ—Ç—Ä—ñ–±–Ω—ñ. –¶–µ –º–æ–∂–Ω–∞ –æ–ø—Ç–∏–º—ñ–∑—É–≤–∞—Ç–∏, –¥–æ–¥–∞–≤—à–∏ –ø—Ä–∞–ø–æ—Ä–µ—Ü—å, —è–∫–∏–π –ø—Ä–∏–ø–∏–Ω—è—î –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∞–ª–≥–æ—Ä–∏—Ç–º—É, —è–∫—â–æ –ø–µ—Ä–µ—Å—Ç–∞–Ω–æ–≤–æ–∫ –Ω–µ –±—É–ª–æ. –¢–æ–º—É —É –Ω–∞–π–∫—Ä–∞—â–æ–º—É –≤–∏–ø–∞–¥–∫—É —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å –±—É–¥–µ ùëÇ(ùëõ).</p></li>
    <li>–°–µ—Ä–µ–¥–Ω—ñ–π –≤–∏–ø–∞–¥–æ–∫ (O(n¬≤)):<p>–£ —Å–µ—Ä–µ–¥–Ω—å–æ–º—É –≤–∏–ø–∞–¥–∫—É –∞–ª–≥–æ—Ä–∏—Ç–º —Ç–∞–∫–æ–∂ –º–∞—î —Å–∫–ª–∞–¥–Ω—ñ—Å—Ç—å ùëÇ(ùëõ2), –æ—Å–∫—ñ–ª—å–∫–∏ –¥–ª—è –≤–ø–æ—Ä—è–¥–∫—É–≤–∞–Ω–Ω—è –µ–ª–µ–º–µ–Ω—Ç—ñ–≤ —Å–µ—Ä–µ–¥–Ω—ñ–π —á–∞—Å –±—É–¥–µ –±–ª–∏–∑—å–∫–∏–º –¥–æ –Ω–∞–π–≥—ñ—Ä—à–æ–≥–æ –≤–∏–ø–∞–¥–∫—É.</p></li>
</ul>

<p>4. –†–µ–∞–ª—ñ–∑—É–≤–∞—Ç–∏ –Ω–∞ Python –∞–ª–≥–æ—Ä–∏—Ç–º —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è –º–µ—Ç–æ–¥–æ–º –±—É–ª—å–±–∞—à–∫–∏.</p>

<p>5. –î–æ—Å–ª—ñ–¥–∏—Ç–∏ —Ä–µ–∞–ª—å–Ω–∏–π —á–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –æ–±–æ—Ö –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤ –Ω–∞ Python —ñ 
–ø–æ–±—É–¥—É–≤–∞—Ç–∏ –≥—Ä–∞—Ñ—ñ–∫–∏ –∑–∞–ª–µ–∂–Ω–æ—Å—Ç–µ–π ùëáùëñùëõùë†ùëíùëüùë°(ùëõ) —Ç–∞ ùëáùëèùë¢ùëèùëôùëí(ùëõ) –¥–ª—è ùëõ = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 10000], –∑–≥–µ–Ω–µ—Ä—É–≤–∞–≤—à–∏ –≤—ñ–¥–ø–æ–≤—ñ–¥–Ω—ñ —Ü—ñ–ª–æ—á–∏—Å–µ–ª—å–Ω—ñ –º–∞—Å–∏–≤–∏.</p>

```python
import time
import random
import matplotlib.pyplot as plt

def insertion_sort(nums):
    for j in range(1, len(nums)):
        key = nums[j]
        i = j - 1
        while i >= 0 and nums[i] > key:
            nums[i + 1] = nums[i]
            i -= 1
        nums[i + 1] = key

def bubble_sort(nums):
    n = len(nums)
    for i in range(n):
        for j in range(0, n-i-1):
            if nums[j] > nums[j+1]:
                nums[j], nums[j+1] = nums[j+1], nums[j]

sizes = [5, 10, 50, 100, 500, 1000, 2000, 3000, 4000, 5000, 10000, 20000, 50000, 10000]
insertion_times = []
bubble_times = []

for size in sizes:
    array = [random.randint(0, 10000) for _ in range(size)]

    start_time = time.time()
    insertion_sort(array.copy())
    insertion_times.append(time.time() - start_time)

    start_time = time.time()
    bubble_sort(array.copy())
    bubble_times.append(time.time() - start_time)

if len(sizes) == len(bubble_times) == len(insertion_times):
    plt.figure(figsize=(10, 6))
    plt.plot(sizes, insertion_times, label='Insertion Sort', marker='o')
    plt.plot(sizes, bubble_times, label='Bubble Sort', marker='o')
    plt.xlabel('–†–æ–∑–º—ñ—Ä –º–∞—Å–∏–≤—É (n)')
    plt.ylabel('–ß–∞—Å –≤–∏–∫–æ–Ω–∞–Ω–Ω—è (—Å–µ–∫—É–Ω–¥–∏)')
    plt.title('–ü–æ—Ä—ñ–≤–Ω—è–Ω–Ω—è —á–∞—Å—É –≤–∏–∫–æ–Ω–∞–Ω–Ω—è –∞–ª–≥–æ—Ä–∏—Ç–º—ñ–≤ —Å–æ—Ä—Ç—É–≤–∞–Ω–Ω—è')
    plt.legend()
    plt.grid(True)
    plt.xscale('log')
    plt.yscale('log')
    plt.show()
```
